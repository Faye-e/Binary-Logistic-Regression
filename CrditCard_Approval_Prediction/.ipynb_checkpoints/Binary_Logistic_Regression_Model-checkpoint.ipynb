When we are dealing with a data set with lots of features, we can use some techniques to selct those features that are more important and will affect our prediction model greatly to reduce the computational cost.


The steps for feature selection are:

- Apply a tree-based estimation method

We are going to use the **Confusion Matrix** to evaluate our model. A confusion matrix is a table that is often used to describe the performance of a classification model on a set of test data for which the true values are known. Then, we will define a function to plot the confusion matrix.

```python
# from sklearn.matrics we imported the confusion_matrix

conf_matrix = confusion_matrix(y_train, y_pred)
conf_matrix
```

To be able to interpret the above matrix more easily, we are going to present the numbers as a percentage by normalizing the numbers.

```python
conf_matrix_norm_ = conf_matrix/conf_matrix.sum(axis=1).reshape(-1,1)
conf_matrix_norm_
```

Now, we can say that our model predicted the 80% of true zeros, however, 18% of zeros are predicted to be ones.

```python
# function to plot the confusion matrix

def plot_Confusion_matrix(matrix, classes = None):
    
    if classes is not None:
        ax = sns.heatmap(matrix, cmap='Oranges', xticklabels=classes, yticklabels=classes, vmin=0., vmax=1.,
                    annot=True, fmt='0.2f', annot_kws={'size':30}, square=True)
        bottom, top = ax.get_ylim()
        ax.set_ylim(bottom + 0.5, top - 0.5)
    else:
        sns.heatmap(matrix, vmin = 0., vmax = 1.)
        
    plt.title('Confusion Matrix')
    plt.ylabel('True labels')
    plt.xlabel('Predicted labels') 
```

```python
plot_Confusion_matrix(conf_matrix_norm_, classes = lg_model.classes_)
```

There are many useful metrics in logistic regression that we can use to evaluate our model. Depending on the type of problem that we are solving, some metrics become more important to focus on. Bellow, we use some of these metrics to investigate our model.

```python
conf_matrix
```

```python
# True Positives (TP) ,True Negatives (TN), False Positives (FP), False Negatives (FN) 

TP = conf_matrix[0][0]
TN = conf_matrix[1][1]
FP = conf_matrix[1][0]
FN = conf_matrix[0][1]


# Accuracy - it is simply a ratio of correctly predicted observation to the total observations. 
Accuracy = (TP +TN) / (TP + FP + FN + TN)
print("The Accuracy is:", Accuracy)

# Precision - it is the ratio of correctly predicted positive observations to the total predicted positive observations.
Precision = TP / (TP + FP)
print("The Precision is:", Precision)

# Recall(Sensitivity) - it is the ratio of correctly predicted positive observations to the all observations in actual class.
Recall = TP / (TP + FN)
print("The Recall is:", Recall)


# F1 score - F1 Score is the weighted average of Precision and Recall. 
F1_Score = 2*(Recall * Precision) / (Recall + Precision)
print("The f1-Score is: ", F1_Score)


# we obtained these results before by using "classification_report(y_train, y_pred)""
```

The interpretation of above metrics according to our problem is:
- Accuracy: Out of all observation, 79% of the time we predicted the 'good' and 'bad' loans correctly. 
- Precision: Out of the total predicted 'bad' loans, 80% of the time we predicted correctly.
- Recall: Out of the total actual 'bad' loans, 81% of the time we predicted correctly.
- f1-score:Intuitively it is not as easy to understand as accuracy, but F1 is usually more useful than accuracy, especially if you have an uneven class distribution. Accuracy works best if false positives and false negatives have similar cost. If the cost of false positives and false negatives are very different, its better to look at both Precision and Recall.
https://blog.exsilio.com/all/accuracy-precision-recall-f1-score-interpretation-of-performance-measures/

The next very important matric that we have to know when evaluation our machine learning model is the **Cross-Entropy loss** or **Log loss** function, which measures the performance of a classification model whose output is **a probability value between 0 and 1**.
This is the loss function used in (multinomial) logistic regression and extensions of it such as neural networks, defined as the negative log-likelihood of a logistic model that returns `y_pred` probabilities for its training data `y_true`. The log loss is only defined for two or more labels. https://scikit-learn.org/stable/modules/generated/sklearn.metrics.log_loss.html

So, our goal is to minimize this function to obtain **the best prediction probability**. Later, we can use parameter tuning to minimize (reduce) the logaritmic loss.

```python
# Apply the cross-entropy function

# log_loss(y_true, y_pred) -->  y_pred: Predicted probabilities, as returned by a classifiers predict_proba method.
# log loss on training data
print('The Log Loss on training data is:', log_loss(y_train, pred_proba))

# log loss on test data (unseen and untrained data)
pred_proba_t = lg_model.predict_proba(X_test)
print('The Log Loss on training data is:', log_loss(y_test, pred_proba_t))
```

As we can expect, the log loss function has a larger value for test data, because we did not train the model on that data set. In the next section, we are performing Hyper parameter tuning to improve our model.

### 4.3. Hyper Parameter Tuning

Logistic Regression has lots of parameters that we can tune. However, for this problem, we are going to focuse only on the **C parameter**, which is Inverse of regularization strength. Usually, when it comes to Logistic Regression or even Neural Networks, this C-parameter is one of the parameters one needs to know how to tweak.
In the discreption of C value in sklearn documentation, it says that **smaller values specify stronger regularization**.the reason is that, **C** value penalizes the weight on large values so it avoids overfitting and enables much stronger regularization.

- We will loop over parameter C (Inverse of regularization strength).
- Inverse of regularization strength helps to avoid overfitting - it penalizes large values of your parameters
- It also helps to find Global Minimum by moving to better "solutions" from local minimum to global minimum
- The values of C to search should be n-equally-spaced values in log space ranging from 1e-5 to 1e5

```python
# selecting logaritmic values for c
np.geomspace(1e-5, 1e5, num=20) 
```

```python
# loop through C-values

C_values = np.geomspace(1e-5, 1e5, num=20) 
cls_accur = []
logarithmic_loss = []

for c in C_values:
    lg_model_new = LogisticRegression(random_state=20, C = c)
    lg_model_new.fit(X_train, y_train)
    score = lg_model_new.score(X_test, y_test)
    cls_accur.append(score)
    print('The classification accuracy of C_value {} is {}'.format(c, score))
    
    log_loss_new = log_loss(y_test, lg_model_new.predict_proba(X_test))
    logarithmic_loss.append(log_loss_new)
    print('The Log_loss of C_value {} is {}'.format(c, log_loss_new))
    print('')
```

What we want to choose is the C parameter that has the lowest logarithmic loss and the highest classification accuracy. To better interpret this outcome, we put the c_values, logarithmic loss and classification accuracy in a dataframe.

```python
# creat a datafram 

print(C_values.shape)
print(np.array(cls_accur).shape)
print(np.array(logarithmic_loss).shape) 

df_metrics = pd.DataFrame(list(zip(C_values, np.array(cls_accur), np.array(logarithmic_loss))), 
                          columns = ['C_values', 'classification accuracy', 'log_loss'])

# sorting the vaues based on the logarithmic loss, from the lowest value to the largets value
df_metrics_sorted = df_metrics.sort_values('log_loss', ascending = True).reset_index()
df_metrics_sorted
```

We want to choose our C parameter based on the lowest logarithmic loss. Sometimes, the lowest log_loss gives us the highest accuracy, but it's not always the case. Here, our highest accuracy is 0.78 which corresponds to larger log_loss. We will stick with the C parameter that gives us the lowest log_loss, which 0.545559.

In the following cell, we will see another way of choosing the optimal C parameter. Sklearn provides a `LogisticRegressionCV` module which implements Logistic Regression with builtin **cross-validation** to find out the optimal C parameter.

```python
# usingcross-validation with kfold to find the optimal C parameter.

from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, random_state=0, shuffle=True)

lg_model_cv = LogisticRegressionCV(cv = kf, random_state=20, Cs = C_values)
lg_model_cv.fit(X_train, y_train)
print('The classification accuracy is', lg_model_cv.score(X_test, y_test))
log_loss_cv = log_loss(y_test, lg_model_cv.predict_proba(X_test))
print('The logarthmic loss is', log_loss_cv)


# calling the optimal C parameter
print('The optimal C parameter is', lg_model_cv.C_)
```

Let's say we want to track a different metric other than classification accuracy and logarithmic loss. In the following cell, we are going to track the change in the confusion matric for different C parameters.

```python
# tracking the confusion matrix

C_values = np.geomspace(1e-5, 1e5, num=20) 
cls_accur = []
logarithmic_loss = []

for c in C_values:
    lg_model_new = LogisticRegression(random_state=20, C = c)
    lg_model_new.fit(X_train, y_train)
    score = lg_model_new.score(X_test, y_test)
    cls_accur.append(score)
    print('The classification accuracy of C_value {} is {}'.format(c, score))
    
    log_loss_new = log_loss(y_test, lg_model_new.predict_proba(X_test))
    logarithmic_loss.append(log_loss_new)
    print('The Log_loss of C_value {} is {}'.format(c, log_loss_new))
    
    # plotting the confusion matrix to see if the 
    y_pred = lg_model_new.predict(X_test)
    conf_matrix = confusion_matrix(y_test, y_pred)
    conf_matrix_norm_ = conf_matrix/conf_matrix.sum(axis=1).reshape(-1,1)
    plot_Confusion_matrix(conf_matrix_norm_, classes = lg_model_new.classes_)
    plt.show()
    # Classification Report
    print(classification_report(y_test, y_pred))
    print('')
```
